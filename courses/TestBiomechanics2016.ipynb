In a long jump, Carl Lewis registered an horizontal velocity off the board of $10.0\,m/s$, with an angle of $18.7^o$ with horizontal plane. Knowing that the center mass position in the beginning was $0.8\,m$ above and $0.10\,m$ in front of the board. Consider $x$ as the horizontal direction and $y$ as the vertical direction, $g$ as the gravity acceleration equal to $10\,m/s^2$. Calculate:  
 a) The movement equations describing the position, velocity and acceleration in the sagittal plane during the jump.  
 b) Maximum height and distance reached, in the horizontal, by the mass center.  
 
<div class='center-align'><figure><img src="carl_lewis.png" width="240" style="float:right; margin: 0 0 10 0px;"/></figure></div>

Data:

Initial velocity, $ v_{0x} = 10\,m/s $  
Take-off angle, $ \theta_0 = 18.7^o; \cos(18.7^o) \approx 0.947 $  
Initial position $ x_{0}=0.10\,m$; $y_{0}=0.8\,m $  
Gravity acceleration, $ g=10\,m/s^2 $ (but points downwards, it's negative)

$$ \cos\theta=\dfrac{10}{v_{0}} $$

$$ v_{0} \approx 10.557 \,m/s $$

$$ v_{0y} = 10.557\sin(18.7^o) \approx 3.385 \,m/s $$

**Movement equations**:

Position:

$$ x(t)=0.10+10t\,[m]$$
$$ y(t)=0.80+3.385t-5t^2\,[m]$$

Velocity:

$$ v_{x}(t)=10\,m/s$$
$$ v_{y}(t)=3,385-10t\,[m/s]$$

Acceleration: 

$$ a_{x}(t)=0\,m/s^2$$
$$ a_{y}(t)=-10\,m/s^2$$

**Maximum height**:

$$ y_{max} = y_0 + h $$

$$h=\frac{v_{0}^2\sin^2(\theta_0)}{2g}$$

$$h=\frac{(10.557)^2\sin^2(18.7)}{2(10)}$$

$$h \approx 0.572\,m$$

$$ y_{max} = 0.8 + 0.572 = 1.372\,m$$

**Jump distance**:

$$ d_{jump} = d_{take-off} + d_{flight} + d_{landing} $$

$$ d_{take-off} = 0.1\,m $$

$$d_{flight+landing}=\dfrac{v_0\cos(\theta_0)\left(v_{0}\sin(\theta_0)+\sqrt{v_{0}^2\sin^2(\theta_0)+2h_{0}g}\right)}{g}$$

$$d_{flight+landing}=\dfrac{(10.557)\cos(18.7)[(10.557)sin(18.7)+\sqrt{(10.557)^2\sin^2(18.7)+2(0.8)(10)}]}{10}$$

$$d_{flight+landing} \approx 8.616\,m$$

$$d_{jump} = 8.716\,m$$ 

2. A diver jumps from a 10-meter tall platform, and reaches the water at a distance of 5-meters in the horizontal after 2.5 seconds. Calculate the following for the mass center of the athlete:  
 a) The Initial velocity and initial angle off the platform.  
 b) The maximum height reached and the velocity when the diver gets into the water.

Data:

Initial height: $y_{0}=10m$  
Horizontal distance reached: $x_{f}=5m$  
Time: $t=2.5s$  

Equations:

In x: 

$$ x_{t} = x_{0} + v_{0}\cos(\theta) t$$
$$ 5 = 0 + v_{0}\cos(\theta) (2.5)$$
$$ \dfrac{5}{2.5} = v_{0}\cos(\theta)$$
$$ 2= v_{0}\cos(\theta)$$

In y:

$$y_{t} = y_{0} + v_{0}\sin(\theta) t - \dfrac{gt^2}{2}$$
$$0 = 10 + v_{0}\sin(\theta) (2.5) - \dfrac{9.81(2.5s)}{2}$$
$$30.65-10=v_{0}\sin(\theta) (2.5)$$
$$8.2625=v_{0}\sin(\theta)$$

Replacing x's $v_{0}$  in y

$$4.13125=\tan(\theta)$$
$$\mathbf{\theta_0 = 76.39^o}$$
$$\mathbf{v_{0}=8.501 \,m/s}$$

For maximum height:

$$h=\dfrac{v_{0}^2\sin^2(\theta)}{2g}$$
$$h=\dfrac{(8.501)^2\sin^2(76.39)}{2(9.81)}$$
$$h=3.48\,m$$
$$y_{max} = y_{0} + h$$
$$\mathbf{y_{max} =13.48\,m}$$

For the velocity when the diver gets into the water:

In x:

$$v_{x}=8.501\cos(76.39)$$
$$v_{x}=2\,m/s$$

In y:

$$v_{y}=8.501\sin(76.39)-(9.81)(2.5)$$
$$v_{y}= -16.2627\,m/s$$
$$v=\sqrt{(v_{x})^2+(v_{y})^2}$$
$$\mathbf{v=16.385\,m/s}$$
